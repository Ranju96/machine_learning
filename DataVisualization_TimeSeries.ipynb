```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Read the data from CSV file
data = pd.read_csv('/content/DailyDelhiClimateTrain.csv')
```

```python


# Group by month and calculate the temperature statistics
monthly_temps = data.groupby(data['date'].str.slice(0, 7)).agg({'humidity': 'mean', 'meantemp': 'mean', 'meantemp': 'mean'})

# Create a stacked bar graph
monthly_temps.plot(kind='bar', stacked=True)
plt.title("Temperature Statistics by Month")
plt.xlabel("Month")
plt.ylabel("Temperature (Celsius)")
plt.show()
```

```python
data = pd.read_csv('/content/DailyDelhiClimateTrain.csv')

# Create a scatter graph
plt.scatter(data['humidity'], data['meantemp'])
plt.title("Minimum vs Maximum Temperature")
plt.xlabel("Minimum Temperature (Celsius)")
plt.ylabel("Maximum Temperature (Celsius)")
plt.show()

```

```python
data = pd.read_csv('/content/DailyDelhiClimateTrain.csv')

# Group by month and calculate the average temperature
monthly_temps = data.groupby(data['date'].str.slice(0, 7))['meantemp'].mean()

# Create a bar graph
plt.bar(monthly_temps.index, monthly_temps)
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (Celsius)")
plt.show()
```
