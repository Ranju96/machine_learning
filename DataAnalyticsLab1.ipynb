```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/Wine.csv')

# Print the first five rows of the dataset
print(data.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Separate the features and labels
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values

# Standardize the features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

# Create a PCA object
pca = PCA(n_components=None)

# Fit the standardized data to the PCA object
pca.fit(X_std)

# Print the explained variance ratio
print(pca.explained_variance_ratio_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the cumulative sum of the explained variance ratio
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot the cumulative sum of the explained variance ratio
plt.plot(cumulative_variance_ratio)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.show()
```

```python
# Project the data onto the first two principal components
X_pca = pca.transform(X_std)[:, :2]

# Plot the data in the new principal component space
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Create a contingency table
cut_points = [0, 100, 200, 300, 400, 500]
data['perimeter_bin'] = pd.cut(data['perimeter_mean'], cut_points)
cont_table = pd.crosstab(data['diagnosis'], data['perimeter_bin'])
```

```python
# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(cont_table)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```

## **chi-square test**

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Load the dataset
data = pd.read_csv('/content/data.csv')
```

```python
# Separate the target variable from the features
X = data.drop(['diagnosis', 'id', 'Unnamed: 32'], axis=1)
y = data['diagnosis']
```

```python
# Create the contingency table
contingency_table = pd.crosstab(y, X['texture_mean'])
```

```python
# Perform the chi-square test
stat, p, dof, expected = chi2_contingency(contingency_table)
```

```python
print('Test statistic:', stat)
print('p-value:', p)
print('Degrees of freedom:', dof)
print('Expected frequencies:', expected)
```

## HR ANALYTICS 

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
data = pd.read_csv('/content/aug_train.csv')
```

```python
# Separate the target variable from the features
X = data.drop(['target'], axis=1)
y = data['target']
```

```python
# Create the contingency table
contingency_table = pd.crosstab(y, X['education_level'])
```

```python
# Perform the chi-square test
stat, p, dof, expected = chi2_contingency(contingency_table)
```

The p-value is 2.49e-06, which is much smaller than the significance level of 0.05. Therefore, we can reject the null hypothesis that the "target" and "education_level" variables are independent. In other words, there is enough evidence to suggest that there is a significant relationship between the "target" and "education_level" variables.

```python
print('Test statistic:', stat)
print('p-value:', p)
print('Degrees of freedom:', dof)
print('Expected frequencies:', expected)
```

### **RFE**

```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
data = pd.read_csv('/content/data.csv')
```

```python
# Separate the target variable from the features
X = data.drop(['diagnosis', 'id', 'Unnamed: 32'], axis=1)
y = data['diagnosis']
```

```python
# Create an instance of the estimator
estimator = RandomForestClassifier()
```

```python
# Create an instance of the RFE object
rfe = RFE(estimator, n_features_to_select=10)
```

```python
# Fit the RFE object to the data
rfe.fit(X, y)
```

```python
# Get a mask of the selected features
mask = rfe.support_

# Get the ranking of all the features
rankings = rfe.ranking_
```

```python
# Get the names of the selected columns
selected_cols = X.columns[mask]

# Print the selected columns and their rankings
for col, rank in zip(X.columns, rankings):
    if col in selected_cols:
        print(col, rank, 'selected')
    else:
        print(col, rank)
```
